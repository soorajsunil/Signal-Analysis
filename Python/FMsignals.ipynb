Generation of Narrowband and Wideband FM signal:

```python
import numpy as np 
import matplotlib.pyplot as plt 

T = 0.01               # sampling period in sec 
Td = 1                  # total signal duration in sec
Fs = 1/T                # sampling frequency in Hz
Ns = Fs*Td              # number of samples calculated for the given time duration 
t = np.arange(0,Ns)*dt  # time vector in sec 

Fc = 250   # freq of carrier signal
Fm = 20    # freq of message signal
beta = 0.5 # frequency modulation index {less than one for narrowband signals, and greater than one for wideband signals}
   
msg = np.sin(2*np.pi*Fm*t)
crr = np.cos(2*np.pi*Fc*t)
nb_fm = crr + beta*msg

beta  = 1.5 
wb_fm = crr + beta*msg

plt.plot(t,crr, 'k', label='carrier signal')
plt.plot(t,msg, 'r', label='message signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.show()

plt.plot(t,nb_fm, label='narrowband FM signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.show()

plt.plot(t,wb_fm, label='wideband FM signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.show()
```
