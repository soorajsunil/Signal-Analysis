```python
import numpy as np 
import matplotlib.pyplot as plt 

# sampling 
fs = 1000 # sampling frequency in Hz
Ts = 1/fs  # sampling period in sec
Ns = 2000  # signal length
t = np.arange(0,Ns)*Ts # time vector 
```

```python
# signal properties 
a = 1    # amplitude 
phi = 0  # phase in rad
f  = 1   # frequency in Hz 

def plot_sinusoid(t, st, ct):
    fig, axs = plt.subplots(2)
    axs[0].plot(t, st)
    axs[0].set_title('sine wave')
    axs[1].plot(t, ct)
    axs[1].set_title('cos wave')

    for ax in axs.flat:
        ax.set(xlabel='Time (s)', ylabel='Amplitude')
        
    # Hide x labels and tick labels for top plots and y ticks for right plots.
    for ax in axs.flat:
        ax.label_outer()
```

```python
# sinusoid
st = a*np.sin(2*np.pi*f*t + phi)
ct = a*np.cos(2*np.pi*f*t + phi)
plot_sinusoid(t, st, ct)
```

```python
# square wave using sign function
st = np.sign(a*np.sin(2*np.pi*f*t + phi))
ct = np.sign(a*np.cos(2*np.pi*f*t + phi))
plot_sinusoid(t, st, ct)
```

```python
# square wave using floor function 
xt = 2*(2*np.floor(f*t) - np.floor(2*f*t)) + 1 
plt.plot(t,xt)
plt.show()
```
