### Useful waveform generation: 

```python
import numpy as np 
import matplotlib.pyplot as plt 

# signal properties 
f = 1    # frequency in Hz 
a = 1    # amplitude 
phi = 0  # phase in rad

# sampling 
T = 0.01  # sampling period 
signal_duration = 10  # in secs 

Fs = 1/T  # sampling frequency in Hertz
tend = signal_duration*Fs  # max time 
t = np.arange(0,tend)*T    # time vector 

def plot_sinusoid(t, st, ct):
    # helper function to plot sinusoids
    fig, axs = plt.subplots(2)
    axs[0].plot(t, st)
    axs[0].set_title('sine')
    axs[1].plot(t, ct)
    axs[1].set_title('cosine')

    for ax in axs.flat:
        ax.set(xlabel='Time (s)', ylabel='Amplitude')
        
    # Hide x labels and tick labels for top plots and y ticks for right plots.
    for ax in axs.flat:
        ax.label_outer()
```

```python
# periodic sinusoids 
st = a*np.sin(2*np.pi*f*t + phi)
ct = a*np.cos(2*np.pi*f*t + phi)
plot_sinusoid(t, st, ct)
```

```python
# periodic square waveform using sign function
st = a*np.sign(np.sin(2*np.pi*f*t + phi))
ct = a*np.sign(np.cos(2*np.pi*f*t + phi))
plot_sinusoid(t, st, ct)
```

```python
# periodic square waveform using floor function 
xt = a*(2*(2*np.floor(f*t) - np.floor(2*f*t)) + 1) 
plt.plot(t,xt)
plt.show()
```

```python
# aperiodic random waveform
w    = 1 # window duration in secs 
wend = w*Fs 

xt = []
for i in range(int(tend/wend)):
    t_i = t[int(i*Fs):int((i+1)*Fs)]
    f_i = f*np.random.randint(0, high=2) # pick random integer between {0,1}
    a_i = a*np.random.randint(-1,high=2) # pick random integer between {-1,0,1}
    xt[int(i*Fs):int((i+1)*Fs)] = a_i*np.sign(np.cos(2*np.pi*f_i*t_i + phi))

plt.plot(t,xt)
plt.show()
```
