### Simple periodic waveforms: 

```python
import numpy as np 
import matplotlib.pyplot as plt 

# signal properties 
f = 1    # frequency in Hz 
a = 1    # amplitude 
phi = 0  # phase in rad

# sampling 
signal_duration = 10  # in secs 
sampling_period = 0.01
sampling_frequency = 1/sampling_period  # in Hertz

tmax = signal_duration*sampling_frequency  # max time 
t = np.arange(0,tmax)*sampling_period # time vector 

def plot_sinusoid(t, st, ct):
    # helper function to plot sinusoids
    fig, axs = plt.subplots(2)
    axs[0].plot(t, st)
    axs[0].set_title('sine')
    axs[1].plot(t, ct)
    axs[1].set_title('cosine')

    for ax in axs.flat:
        ax.set(xlabel='Time (s)', ylabel='Amplitude')
        
    # Hide x labels and tick labels for top plots and y ticks for right plots.
    for ax in axs.flat:
        ax.label_outer()
```

```python
# sinusoids 
st = a*np.sin(2*np.pi*f*t + phi)
ct = a*np.cos(2*np.pi*f*t + phi)
plot_sinusoid(t, st, ct)
```

```python
# square wave using sign function
st = a*np.sign(np.sin(2*np.pi*f*t + phi))
ct = a*np.sign(np.cos(2*np.pi*f*t + phi))
plot_sinusoid(t, st, ct)
```

```python
# square wave using floor function 
xt = a*(2*(2*np.floor(f*t) - np.floor(2*f*t)) + 1) 
plt.plot(t,xt)
plt.show()
```
